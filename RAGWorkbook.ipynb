Define a function to generate vector embeddings

```python
import os
import voyageai

#Specify the embedding model
model = "voyage-finance-2"
voyageai.api_key = os.environ["VOYAGE_API_KEY"] # Explicitly set API key
vo = voyageai.Client()

#Define a function to generate embeddings
def get_embedding(data, input_type = "document"):
  embeddings = vo.embed(
      data, model = model, input_type = input_type
  ).embeddings
  return embeddings[0]
```

Load and split the data.

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

#Load the pdf
Loader = PyPDFLoader("https://investors.mongodb.com/node/12236/pdf")
data = Loader.load()

#Split the data into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=20)
document_chunks = text_splitter.split_documents(data)
```

Convert the data into vector embeddings

```python
#Prepare documents for ingestion
docs_to_insert = [{
    "text": doc.page_content,
    "embedding": get_embedding(doc.page_content)
}for doc in document_chunks]
```
